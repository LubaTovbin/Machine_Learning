```python
def my_hw (N, R, slope, intercept):
# N - number of points, R - range    
    print('target weights: w0 = ',-intercept,'  w1 = ',-slope,'  w2 = ', 1 )
    print('target slope and intercept are ', slope,' and ', intercept)

    # plot target perceptron
    t= np.arange(R + 1)
    f= slope*t + intercept 
    plt.plot(t, f, linestyle='-', color='green', label= 'Target')

    # randomize first coordinate, x1, N points
    x1 = np.array(np.random.randint(R, size= N), dtype = np.float64)

    # randomize second coordinate, x2 
    x2 = np.array(np.random.randint(R, size= N), dtype = np.float64)

    # create an array to hold training values of y's for each pair(x1,x2)
    y= np.empty(N, dtype= np.float64)
    for i in range(N):
        if x2[i] > slope*x1[i] + intercept:
            y[i]= 1.0
            plt.scatter(x1[i], x2[i], marker='o', color='b')
        else:
            y[i]= -1.0
            plt.scatter(x1[i], x2[i], marker='x', color='r')
    plt.xlabel('x1')
    plt.ylabel('x2')

    # PLA
    # initial weigths
    w0 = w1 = w2 = 1.0
    #number of iterations, maximum iterations allowed is 10,000
    k = 0
    # flag to indicate that no missclassified points left
    done = False
    
    while done == False and k < 10000:       
        for i in np.random.permutation(N):
            done = True
            s = w0 + w1*x1[i] + w2*x2[i]
            if  s*y[i] <= 0:                        
                w0= w0 + y[i]
                w1= w1 + y[i]*x1[i]
                w2= w2 + y[i]*x2[i]
                done = False
                break        
        k= k + 1
               
    print('number of interations', k)
    print('learned wieghts: ''w0=', w0,'  w1=', w1, '  w2=', w2)

    # final hypothesis, g
    g= -(w1*t)/w2 - w0/w2
    plt.plot(t, g, linestyle='--', color='black', label= 'Learned')
    plt.legend(loc= 'best')
    
    print('learned slope and intercept are ', -(w1/w2),' and ', -(w0/w2) )
    
```

```python
# HW1 Problem 8 a,b
my_hw(20, 100, -0.5, 60.0)
```

Comments on 8b: 
The final hypothesis, g, is close to the target function, f. No point is misclassified.
